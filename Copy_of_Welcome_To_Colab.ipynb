```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample Data (Replace this with actual dataset)
data = {
    'App': ['App1', 'App2', 'App3', 'App4', 'App5', 'App6', 'App7', 'App8'],
    'Category': ['Game', 'Utility', 'Finance', 'Game', 'Health', 'Finance', 'Utility', 'Health'],
    'Installs': [5000, 12000, 8000, 15000, 7000, 9000, 11000, 6000],
    'Revenue': [20000, 34000, 27000, 45000, 19000, 23000, 31000, 18000],
    'Type': ['Paid', 'Paid', 'Paid', 'Paid', 'Paid', 'Paid', 'Paid', 'Paid']
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Filter only paid apps
df_paid = df[df['Type'] == 'Paid']

# Set color palette
palette = sns.color_palette("husl", len(df_paid['Category'].unique()))

# Create scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_paid, x='Installs', y='Revenue', hue='Category', palette=palette, s=100, edgecolor='black'
)

# Add trendline
z = np.polyfit(df_paid['Installs'], df_paid['Revenue'], 1)
p = np.poly1d(z)
plt.plot(df_paid['Installs'], p(df_paid['Installs']), "r--", label="Trendline")

# Labels and title
plt.xlabel('Number of Installs')
plt.ylabel('Revenue ($)')
plt.title('Revenue vs. Installs for Paid Apps')
plt.legend(title="Category")
plt.grid(True)

# Show plot
plt.show()
```
